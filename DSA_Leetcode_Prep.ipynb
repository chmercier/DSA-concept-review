**Binary Search**

```python
def BinarySearch(nums, key):
    low = 0
    high = len(nums) - 1

    while low <= high:
        mid = (high + low) // 2

        if nums[mid] == key:
            return mid

        elif nums[mid] > key:
            high = mid - 1

        elif nums[mid] < key:
            low = mid + 1

    return -1
```

**Linear Search**

```python
def LinearSearch(arr, key):
    index = 0

    for element in arr:
        if element == key:
            return index

        index += 1

    return -1
```

**Two Sum**

```python
def twoSumMedium(arr, target):
    low = 0
    high = len(arr) - 1

    arr.sort()

    while low < high:
        sum = arr[high] + arr[low]

        if sum < target:
            low += 1

        elif sum > target:
            high -= 1

        elif sum == target:
            return arr[high], arr[low]

    return "no match"

def twoSumOptimalDict(num, target):
    match_dict = {}

    for i in range(len(num)):
        match = target - num[i]

        if match in match_dict:
            return match_dict[match], i

        match_dict[num[i]] = i

    return "no match"


def twoSumOptimal(arr, target):
    match_set = set()

    for element in arr:
        match = target - element

        if match in match_set:
            return match, element

        match_set.add(element)

    return "No match"
```

**Single Element in Array**

```python
def singleElementInArray(arr):
    for index in range(0, len(arr) - 1, 2):
        if arr[index] != arr[index + 1]:
            return arr[index]

    return arr[-1]

def singleElementInArrayBinary(arr):
    low = 0
    high = len(arr) - 1

    while low < high:
        mid = (low + high) // 2

        if mid % 2 == 0:
            mid -= 1

        if arr[mid] == arr[mid - 1]:
            low = mid + 1

        elif arr[mid] != arr[mid - 1]:
            high = mid - 1

    return arr[low]
```

**Count Ones**

```python
def CountOnes(arr):
    count_ones = 0

    for element in arr:
        if element == 1:
            count_ones += 1
        else:
            break

    return count_ones

def CountOnesBinary(arr):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (high + low) // 2

        if arr[mid] == 0:
            high = mid - 1

        elif arr[mid + 1] != 1:
            return mid + 1

        elif arr[mid] == 1:
            low = mid + 1

    return 0
```
